```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
file_path = "../eval/llama3_evol_exp_lr3e-5_peft_constant_r128_a128/error_breakdown"

validation_loss = []
for i in range(500,3200,100):
    fn = f"{file_path}/{i}.csv"
    data = pd.read_csv(fn)
    success_rate = data.iloc[3]["error_count"] / 80 * 100
    validation_loss.append(success_rate)
    
training_steps = pd.read_csv("llama3_ri_exp_lr3e-5_peft_constant_training_loss.csv")

training_loss = training_steps["llama3_ri_exp_lr3e-5_peft_constant - train/loss"].tolist()
```

```python
plt.figure()

plt.plot(range(500,3200,100), validation_loss, label="Validation Loss")
plt.xlabel("Training Steps")
plt.ylabel("Pass@1 Score (%)")
plt.figure()
plt.plot(range(0,3125,5), training_loss, label="Validation Loss")
plt.xlabel("Training Steps")
plt.ylabel("Training Loss")
```
